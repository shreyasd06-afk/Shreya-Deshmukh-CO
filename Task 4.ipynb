```python
import matplotlib.pyplot as plt

# Data
years = list(range(2010, 2020))
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
runs = [200, 400, 600, 900, 1300, 1500, 1600, 1700, 1800, 1850]  # Tendulkar

# Get 10 available matplotlib styles
styles = plt.style.available[:10]  # Select any 10 styles

# Plot for each style
for style_name in styles:
    plt.style.use(style_name)  # Apply the style
    plt.figure(figsize=(8, 5))

    # Plotting the lines
    plt.plot(years, kohli, color='orange', linestyle='--', linewidth=3, marker='o', label="Kohli")
    plt.plot(years, sehwag, color='green', linestyle='-.', linewidth=2, marker='s', label="Sehwag")
    plt.plot(years, runs, color='blue', linestyle='-', linewidth=2, marker='^', label="Tendulkar")

    # Titles and labels with correct style name
    plt.title(f"Performance Comparison - Style: {style_name}")
    plt.xlabel("Year")
    plt.ylabel("Runs Scored")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2010, 2012, 2014, 2016, 2018]

sachin = [1200, 1000, 700, 0, 0]    
sehwag = [1000, 1100, 900, 400, 0]
kohli = [300, 800, 1100, 1300, 1600]

bar_width = 0.25


x = np.arange(len(years))
x1 = x
x2 = x + bar_width
x3 = x + 2 * bar_width

plt.figure(figsize=(10, 6))
plt.bar(x1, sachin, width=bar_width, label="Sachin", color='blue')
plt.bar(x2, sehwag, width=bar_width, label="Sehwag", color='green')
plt.bar(x3, kohli, width=bar_width, label="Kohli", color='orange')


plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison: Sachin vs Sehwag vs Kohli")
plt.xticks(x + bar_width, years)
plt.legend()
plt.tight_layout()
plt.grid(axis='y')

plt.show()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]

sachin_5yr = 3500
sehwag_5yr = 3200
kohli_5yr = 4000

runs = [sachin_5yr, sehwag_5yr, kohli_5yr]

plt.figure(figsize=(8, 4))
plt.barh(players, runs, color=["red", "pink", "orange"])
plt.xlabel("Total Runs in First 5 Years")
plt.title("Debut 5-Year Performance Comparison")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

os_labels = ['Android', 'iOS', 'KaiOS', 'Windows Phone', 'Blackberry', 'Others']
market_share = [72, 26, 0.5, 0.3, 0.1, 1.1]
colors = ['red', 'orange', 'green', 'blue', 'black', 'purple']


plt.figure(figsize=(7, 7))
plt.pie(market_share, labels=os_labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Mobile OS Market Share (Pie Chart)")
plt.axis('equal')
plt.tight_layout()
plt.show()


plt.figure(figsize=(9, 5))
plt.barh(os_labels, market_share, color=colors)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

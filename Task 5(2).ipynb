```python
#(1)Plot a Line Chart: Plot y = 3x + 2 for x in the range  0 to 10. Set the 
#x-axis and y-axis labels,  and give the plot a title.
```

```python
#(1)
import numpy as np

x = np.arange(0, 11)  
y = 3 * x + 2         


plt.plot(x, y, marker='o', color='b')


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart for y = 3x + 2')


plt.grid(True)


plt.show()
```

```python
#(3)
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 25, 10, 30, 20]

plt.bar(categories, values, color='skyblue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Annotations')

for i, val in enumerate(values):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=10)
    
plt.grid(axis='y')

plt.show()
```

```python
#(4)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)  
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)

plt.plot(x, y1, color='blue', linestyle='-', label='sin(x)')     # solid line
plt.plot(x, y2, color='red', linestyle='--', label='cos(x)')     # dashed line
plt.plot(x, y3, color='green', linestyle=':', label='exp(x)')    # dotted line

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Functions on One Plot')
plt.legend()

plt.grid(True)

plt.show()
```

```python
#(6)
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)

hist_data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 1 - Line plot
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Plot')

# 2 - Bar chart
axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title('Bar Chart')

# 3 - Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='green', s=50)
axs[1, 0].set_title('Scatter Plot')

# 4 - Histogram
axs[1, 1].hist(hist_data, bins=20, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram')

fig.suptitle('2x2 Grid of Different Plots', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.96])

plt.show()
```

```python
#(7)
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

product_A = [5, 6, 7, 8, 7, 8, 9, 10, 9, 8, 7, 6]
product_B = [2, 3, 4, 3, 4, 5, 4, 5, 6, 5, 4, 3]
product_C = [1, 2, 1, 2, 3, 2, 3, 2, 3, 4, 3, 2]

plt.stackplot(months, product_A, product_B, product_C, 
              labels=['Product A', 'Product B', 'Product C'], 
              colors=['skyblue', 'lightgreen', 'lightcoral'])

plt.xlabel('Month')
plt.ylabel('Sales (in units)')
plt.title('Cumulative Sales of 3 Products Over 12 Months')

plt.legend(loc='upper left')

plt.show()
```

```python
#(8)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'Math': np.random.randint(50, 100, 20),
    'Science': np.random.randint(40, 95, 20),
    'English': np.random.randint(45, 90, 20)
}
df = pd.DataFrame(data)

print(df)

plt.boxplot([df['Math'], df['Science'], df['English']], labels=['Math', 'Science', 'English'])

plt.ylabel('Scores')
plt.title('Exam Scores Distribution')

plt.grid(True)

plt.show()
```

```python
#(9)
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [35, 25, 20, 20]  # percentages

largest_index = market_share.index(max(market_share))


explode = [0.1 if i == largest_index else 0 for i in range(len(companies))]

plt.pie(market_share, labels=companies, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis('equal')

plt.title('Market Share of Companies')

plt.show()
```

```python
#(10)
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(10, 50, size=10)

plt.plot(x, y1, label='Line 1', marker='o')
plt.plot(x, y2, label='Line 2', marker='s')

plt.gca().invert_xaxis()

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Line Plot with Custom Axis & Legend')

plt.grid(True)

plt.tight_layout()

plt.show()
```
